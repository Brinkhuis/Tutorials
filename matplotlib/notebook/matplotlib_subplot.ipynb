# Subplots

#### import modules

```python
import sys
import numpy as np
from matplotlib import pyplot as plt
```

#### version info

```python
print('python     {}'.format(sys.version.split(' ')[0]))
print(f'numpy      {np.__version__}')
print('matplotlib {}'.format(sys.modules['matplotlib'].__version__))
```

#### basic plots

```python
x = np.arange(-5, 5, .1)
```

```python
plt.plot(x, x**2, c='blue')
plt.show()
```

```python
plt.plot(x, x**3, c='red')
plt.show()
```

```python
plt.plot(x, x**2, c='blue')
plt.plot(x, x**3, c='red')
plt.show()
```

#### subplots, rows

```python
f, ax = plt.subplots()
ax.plot(x, x**2, c='blue')
ax.plot(x, x**3, c='red')
plt.show()
```

```python
f, ax = plt.subplots(2)
ax[0].plot(x, x**2, c='blue')
ax[1].plot(x, x**3, c='red')
plt.show()
```

```python
f, ax = plt.subplots(3, sharex=True)
ax[0].plot(x, x**2, c='blue')
ax[1].plot(x, x**3, c='red')
ax[2].plot(x, x**4, c='green')
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
ax1.plot(x, x**2, c='blue')
ax2.plot(x, x**3, c='red')
ax3.plot(x, x**4, c='green')
plt.show()
```

```python
subplt = 4
colors = ['blue', 'red', 'green', 'orange']

f, ax = plt.subplots(subplt, sharex=True, figsize=(5, 10))
for num in range(subplt):
    ax[num].plot(x, x**(num+2), c=colors[num], linewidth=3)
plt.show()
```

#### subplots, columns

```python
f, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, x**2, c='blue')
ax[1].plot(x, x**3, c='red')
plt.show()
```

```python
f, ax = plt.subplots(nrows=1, ncols=2, sharey=True)
ax[0].plot(x, x**2, c='blue')
ax[1].plot(x, x**3, c='red')
plt.show()
```

```python
f, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
ax[0].plot(x, x**2, c='blue')
ax[1].plot(x, x**3, c='red')
ax[2].plot(x, x**4, c='green')
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, figsize=(15, 5))
ax1.plot(x, x**2, c='blue')
ax2.plot(x, x**3, c='red')
ax3.plot(x, x**4, c='green')
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, figsize=(15, 5))

ax1.plot(x, x**2, c='blue')
ax1.set_title('$f(x) = x^2$', color='blue', fontsize=14)

ax2.plot(x, x**3, c='red')
ax2.set_title('$f(x) = x^3$', color='red', fontsize=14)

ax3.plot(x, x**4, c='green')
ax3.set_title('$f(x) = x^4$', color='green', fontsize=14)
plt.show()
```

#### subplots, rows and columns

```python
f, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
ax[0, 0].plot(x, x**2, c='blue')
ax[0, 1].plot(x, x**3, c='red')
ax[1, 0].plot(x, x**4, c='green')
ax[1, 1].plot(x, x**5, c='orange')
plt.show()
```

```python
f = plt.figure(figsize=(12, 12))
f.suptitle('Subplots', fontsize=24)

plt.subplot(2, 2, 1)
plt.plot(x, x **2, c='red')
plt.title('$f(x) = x^2$', color='red', fontsize=18)

plt.subplot(2, 2, 2)
plt.plot(x, x ** 3, c='blue')
plt.title('$f(x) = x^3$', color='blue', fontsize=18)

plt.subplot(2, 2, 3)
plt.plot(x, x ** 4, c='green')
plt.title('$f(x) = x^4$', color='green', fontsize=18)

plt.subplot(2, 2, 4)
plt.plot(x, x ** 5, c='orange')
plt.title('$f(x) = x^5$', color='orange', fontsize=18)

plt.show()
```

```python
f = plt.figure(figsize=(12, 12))
f.suptitle('Even more subplots!', fontsize=24)

plt.subplot(3, 2, 1)
plt.plot(x, x **2, c='red')

plt.subplot(3, 2, 2)
plt.plot(x, x ** 3, c='blue')
plt.axis('off')

#plt.subplot(3, 2, 3)
#plt.plot(x, x ** 4, c='green')

plt.subplot(3, 2, 4)
plt.plot(x, x ** 5, c='orange')

plt.subplot(3, 2, 5)
plt.scatter(x, np.sin(x), c='purple', marker='o', s=10)

plt.subplot(3, 2, 6)
plt.scatter(x, np.cos(x*2), c='darkgrey', marker='o', s=15)

plt.show()
```
